```python
data_f = vk.users.search(city_id=31, sex='2', fields=['first_name', 'last_name', 'sex', 'bdate'], count = 500, sort=1)['items']
filtered_women = [user for user in data_f if 'bdate' in user and user['bdate'].count('.') == 2]
result_users = list()
for user in filtered_women:
    result_users.append(user)
result_users
```
